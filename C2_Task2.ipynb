## Discretization

### By Age

```python
# dataAge will have original data with bins for AGE and LIMIT_BAL

dataAge = credit
dataAge['age_int'] = pd.cut(dataAge.AGE, [20,30,40,50,60,70,80])
```

### By Limit_Bal

```python
dataAge['limit_int'] = pd.cut(dataAge.LIMIT_BAL,[9999,100000,200000,300000,400000,500000,600000,700000,800000,900000,1000000])
```

```python
# Linear plot of Limit_bal vs. Payment with Default customers
sns.lmplot('default payment next month_default','LIMIT_BAL', data=dum_credit)
plt.title('Trend between Payment and the Limit credit', fontsize=15, pad=10)
plt.show()
```

There seems to be a general linear trend between Payment and the Limit credit. The plot shows that the higher the limit is, the less chance he/she would default.
